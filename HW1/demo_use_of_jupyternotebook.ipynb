```python
# make simple Matplotlib figures to be saved in PNGs
# Configure Matplotlib
plt.rcParams['font.family'] = 'DejaVu Sans'  # or another Unicode-compatible font
plt.rcParams['axes.unicode_minus'] = False   # Disable Unicode minus sign

# Generate data
x = np.linspace(-10, 10, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Save the plot as PNG
plt.savefig('sine_wave.png')

# Optional: Display the plot
plt.show()
```
